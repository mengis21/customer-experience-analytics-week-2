```python
import pandas as pd
import psycopg2
```

```python
# Connect to PostgreSQL

conn = psycopg2.connect(
    host="localhost",
    database="restaurant_reviews",
    user="admin",
    password="admin123"
)
```

```python
# Define a Helper Function to Run SQL and Return DataFrame

def run_query(sql):
    return pd.read_sql(sql, conn)
```

```python
sql = "SELECT * FROM reviews;"

run_query(sql)
```

```python
sql = "SELECT * FROM restaurants;"

run_query(sql)
```

```python
# Count reviews per restaurant

sql = """
SELECT r.restaurant_name, COUNT(*) AS total_reviews
FROM reviews v
JOIN restaurants r ON r.restaurant_id = v.restaurant_id
GROUP BY r.restaurant_name;
"""

run_query(sql)
```

```python
# Average rating per restaurant

sql = """
SELECT r.restaurant_name, ROUND(AVG(v.rating), 2) AS avg_rating
FROM reviews v
JOIN restaurants r ON r.restaurant_id = v.restaurant_id
GROUP BY r.restaurant_name;
"""

run_query(sql)
```
